```python
from pathlib import Path
import numpy as np
from PIL import Image
from scipy.fft import dctn
```

Save absolute values of DCT coefficients (Training data)

```python
input_dir = Path("img50x50")
output_dir = Path("abs_dct50x50")
output_dir.mkdir(exist_ok=True)

for i in range(10001, 10101):
    img_path = input_dir / f"image_{i:05d}.jpg"
    
    img = Image.open(img_path).convert("L")
    arr = np.asarray(img, dtype=np.float64)

    # 2D orthonormal DCT
    dct_arr = dctn(arr, type=2, norm="ortho")
    abs_dct = np.abs(dct_arr)
    # Save absolute values of DCT coefficients
    np.save(output_dir / f"abs_dct_{i:05d}.npy", abs_dct)

print(f"Saved files to: {output_dir.resolve()}")
```
