```python
import requests
from pathlib import Path
```

```python
out = Path("img50x50")
out.mkdir(exist_ok=True)

for i in range(1, 10001):
    url = f"https://picsum.photos/seed/random_{i}/50/50?grayscale"
    r = requests.get(url, timeout=30)
    r.raise_for_status()
    (out / f"image_{i:05d}.jpg").write_bytes(r.content)
```
